```python
dic={
    'a':1,
     'b':2,
     'c':3,
     'd':4,
     'e':1,
     'f':2,
     'g':3,
     'h':4,
     'i':1,
     'j':2,
     'k':3,
     'l':4,
     'm':1,
     'n':2,
     'o':3,
     'p':4,
     'q':1,
     'r':2,
     's':3,
     't':4,
     'u':1,
     'v':2,
     'w':3,
     'x':4,
     'y':1,
     'z':2
}

n=int(input("Enter no of matches to play : "))

score=0
score2=0
for i in range(n):
  if i==0:
    str=input("P1 : Enter a word : ")

    for x in str:
       score+=dic[x]
    print("start with {}".format(str[-1:]))
    str2=input("P2 : Enter a word : ")
    incr=0
    for x in str2:
      incr+=dic[x]

    if str[-1:] != str2[0:1]:
      score2-=incr
    else:
      score2+=incr
    print("p1 score is {}".format(score))
    print("p2 score is {}".format(score2))
    if n!=1:
      print("start with {}".format(str2[-1:]))
  else:
    str=input("P1 : Enter a word : ")
  
    incr=0
    for x in str:
      incr+=dic[x]

    if str2[-1:] != str[0:1]:
      score-=incr
    else:
     score+=incr
    print("start with {}".format(str[-1:]))
  
    str2=input("P2 : Enter a word : ")
    incr=0
    for x in str2:
      incr+=dic[x]

    if str[-1:] != str2[0:1]:
      score2-=incr
    else:
      score2+=incr
    
    print("p1 score is {}".format(score))
    print("p2 score is {}".format(score2))
    if i<n-1:
      print("start with {}".format(str2[-1:]))

if score==score2:
  print("Score equals...match tied")
elif score>score2:
  print("p1 won by {} points".format(score-score2))
else:
  print("P2 won by {} points".format(score2-score))
```
